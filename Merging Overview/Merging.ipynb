### The type of data merge requires an understanding of what you are looking for

```python
import pandas as pd
```

```python
regularSeasonData = pd.read_csv("3YearRegularSeasonData.csv")
playoffData = pd.read_csv("3YearPlayoffData.csv")
```

```python
print(regularSeasonData)
print(playoffData)
```

# **1. Left Join (Most Common Join)**

### Think: “Keep everything I care about in the left table, and add whatever matches from the right table.”

<div align="center">
    <img src="LeftJoin.png" width="300">
</div>

```python
merged_df = pd.merge(regularSeasonData, playoffData, on='Team', how='left')
merged_df
```

# **2. Right Join (Inverse of Left Join)**

### Think: “Keep everything I care about in the right table, and add whatever matches from the left table.”

<div align="center">
    <img src="RightJoin.png" width="300">
</div>

```python
merged_df = pd.merge(regularSeasonData, playoffData, on='Team', how='right')
merged_df
```

# **3. Inner Join**

### Think: “Keep only the rows that exist in both tables; throw away everything else.”

<div align="center">
    <img src="InnerJoin.png" width="300">
</div>

```python
merged_df = pd.merge(regularSeasonData, playoffData, on='Team', how='inner')
merged_df
```

# **4. Outer Join (Also known as Full Join)**

### Think: “Keep everything from both tables; if there’s no match, leave a gap (NULL).”

<div align="center">
    <img src="OuterJoin.png" width="300">
</div>

```python
merged_df = pd.merge(regularSeasonData, playoffData, on='Team', how='outer')
merged_df
```
