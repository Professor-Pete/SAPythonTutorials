## **Print Statements**

### *Used for outputs and easy debugging. You are telling Python to display something*

```python
#Tell Python to display Hello World

print("Hello World")
```

```python
#Tell Python to display the number four

print(4)
```

## **Python stores data as variables**

```python
# Store the value 'Hello World' into x
# Start with the new variable being created

x = 'Hello World'
print(x)

y = 4
print(y)
```

## **Data Types are different methods of storing data**

### *Data Types include integers, strings, floats, lists, dictionaries, booleans*

```python
# Type() is a method used to check data type
# As you can see, after storing the value into x, Python determines the data types

# Print the data type of the variable 'u'
u = 1
print(type(u))

# Print the data type of the variable 'v'
v = 'John'
print(type(v))

# Print the data type of the variable 'w'
w = 2.4
print(type(w))

# Print the data type of the variable 'x'
x = ['John', 'James']
print(type(x))

# Print the data type of the variable 'y'
y = {'Name': 'John', "Name2": "Matthew"}
print(type(y))

# Print the data type of the variable 'z'
z = True
print(type(z))
```

### *Once you re-assign a variable, the new value replaces the old value*

```python
# # # Re-assgining x to 2

x = 1
x = 2
x = 1

print(x)
```

## **You can add variables together using mathematical operations**
### *This includes integers and strings*

```python
# Re-assgining x to 2

w = 1
x = 2
print(w + x)

y = 'Hello'
z = 'World'
print(y + z)
```

### *Always assign variables to new values created*
#### *Start with the new variable being created*

```python
# Adding w and x and assigning to a

a = w + x

# Adding y and z and assigning to b
b = y + z

print(a)
print(b)
```

## **IF Statements**

```python
# If statements are essential in Python and one of the most basic building blocks of writing code
# Using multiple if's allows for multiple outputs

x = 5

if x == 5:
    print('x is equal to 5')
if x > 5:
    print('x is greater than 5')
if x < 5:
    print('x is less than 5')
if x >= 5:
    print('x is greater than or equal to 5')
if x <= 5:
    print('x is less than or equal to 5')
if x != 5:
    print('x is not equal to 5')
```

```python
# Use elif and else when having only one option
# Using elif and else only gives one output

x = 10

if x > 10:
    print('x is greater than 10')
else:
    print('x is 10')
```
