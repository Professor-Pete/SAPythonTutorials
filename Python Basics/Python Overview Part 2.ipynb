## **Lists, Indexing, and Slicing**

```python
# Lists are indicated by brackets or []

fruits = ['oranges', 'apples', 'pears']
print(fruits)
```

```python
# Python has built-in functions to sort lists 

fruits.sort()
fruits
```

```python
# You can sort both ways
fruits.sort(reverse=True)
fruits
```

```python
# You can select certain parts of a list with an index
# Python starts counting from 0, not 1

print(fruits[0])
print(fruits[1])
print(fruits[2])
```

```python
# For long lists, you can use slicing to select certain values

print(fruits[:1])
print(fruits[:3])
```

```python
# You can also slice starting from the last value

print(fruits[2:])
```

## **Iterations (AKA Loops)**

```python
# Using a for loop to iterate through a range of numbers

for i in range(5):
    print("This is iteration", i)
```

### *Using loops is one of the biggest advantages of coding. It enables the automation of re-occuring tasks*

```python
# This loop goes through my list of fruits, and displays which ones I like or don't like based on conditions

fruits = ["apple", "banana", "cherry", "date"]

# Using a for loop to print each fruit
for fruit in fruits:
    if fruit in ['banana', 'cherry']:
        print(f"I like {fruit}s!")
    elif fruit in ['apple', 'date']:
        print(f"I don't like {fruit}s!")
```

## **Dictionaries**

### *Dictionaries are lists but with keys attached to values*

```python
my_dict = {"Boys": ['John', 'James'], "Girls": ['Jennie', 'Julie']}
print(my_dict)
```

```python
# You can access values by using the keys

print(my_dict['Boys'])
print(my_dict['Girls'])
```
