## ***Python Pandas Package Overview - Part 2***

```python
import pandas as pd
```

## *Read in our file*

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
fortune
```

## *See size of unique data*

```python
fortune['Sector'].nunique()
```

## *Using groupby*

```python
sectors = fortune.groupby("Sector")
sectors
```

## *Analyzing columns*

```python
sectors.size()
```

## *Using Mathemtical Methods*

```python
sectors['Revenue'].max()
sectors['Revenue'].min()
```

## *Using .agg() method*

```python
# You can use .agg() for sum, min, max, mean, etc...

sectors.agg ({'Revenue':'sum','Profits':'sum','Employees':'mean'})
```

## *Filtering groupby DataFrame*

```python
sector_revenue = sectors["Revenue"].sum()
sector_revenue[sector_revenue > 1_000_000]
```

## *Using .get_group()*

```python
sectors.get_group('Energy')
```

```python
energy = sectors.get_group('Energy')
print(energy['Revenue'].sum())
```
