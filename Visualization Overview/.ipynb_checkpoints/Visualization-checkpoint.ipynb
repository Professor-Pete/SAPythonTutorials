# ***Visualization Overview***

```python
import pandas as pd
```

```python
df = pd.read_csv("playerData.csv")
df
```

## **Bar Chart**

```python
import matplotlib.pyplot as plt
```

```python
# Size of the chart
plt.figure(figsize=(10, 6))

# Example different colors for each player
# colors = ['red', 'green', 'blue', 'orange', 'purple']
# Highlight one player
# colors = ['skyblue', 'orange', 'skyblue', 'skyblue', 'skyblue']

# Create a bar chart
plt.bar(df['Player'], df['Batting Average'], color='skyblue', label='Batting Average')

# Add title and labels
plt.title('Player Batting Averages')
plt.xlabel('Player')
plt.ylabel('Batting Average')
plt.ylim(0.0, 0.35)  # Y-axis range for clarity
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Only horizontal grid lines
plt.legend()
plt.show()
```

## **Line Chart**

```python
# Size of the chart
plt.figure(figsize=(10, 6))

# Define the chart
plt.plot(df['Player'], df['Home Runs'], marker='o', linestyle='-', color='blue', label='Home Runs')

# Add title and labels
plt.title('Player Home Runs')
plt.xlabel('Player')
plt.ylabel('Home Runs')
plt.ylim(0, 60)  # Y-axis range for clarity
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

# **Scatter Plot**

```python
import seaborn as sns
```

```python
# Size of the chart
plt.figure(figsize=(10,6))

# Define the chart
sns.scatterplot(x='WAR', y='Home Runs', hue='Player', data=df, s=100)

# Add titles and labels
plt.title("WAR vs Home Runs")
plt.xlabel("WAR")
plt.ylabel("Home Runs")
plt.legend(title='Player')
plt.grid(True)
plt.show()
```
